## json簡易操作

*  + 	打開一個文件進行更新(可讀可寫)
*  r 	以只讀方式打開文件
*  r+ 	打開一個文件用於讀寫
*  w 	打開一個文件只用於寫入，如果該文件已存在則打開文件，並從開頭開始編輯，原有內容會被刪除。如果該文件不存在，創建新文件。
*  w+ 	打開一個文件用於讀寫。如果該文件已存在則打開文件，並從開頭開始編輯，原有內容會被刪除。如果該文件不存在，創建新文件。
*  a 	打開一個文件用於追加。如果該文件已存在，新的內容將會被寫入到已有內容之後。如果該文件不存在，創建新文件進行寫入。
*  a+ 	打開一個文件用於讀寫。如果該文件已存在，文件指針將會放在文件的結尾。文件打開時會是追加模式。如果該文件不存在，創建新文件用於讀寫。
*  ab+ 	以二進制格式打開一個文件用於追加。

### 測試一個網址

```python
def scrapwine():
    path = ("chromedriver")
    s = Service(path)
    ua = UserAgent()
    user_agent = ua.random
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('user-agent = user_agent')
    driver = webdriver.Chrome(service=s,chrome_options=chrome_options)
    sec = random.choice([6,7,8,9,10,11])#隨機休息秒數

    #也存成json檔
    filename = "test.json"
    jsonObject = {
        "Link":"nan",
        "winery":"nan",
        "wine_name":"nan",
        "no":"nan",
        "year":"nan",
        "price_id":"nan",
        "country":"nan",
        "region":"nan",
        "wine_type":"nan",
        "grape_id":"nan",
        "rating":"nan",
        "no_ratings":"nan",
        "highlight1":"No highlight1",
        "hightlight2":"No highlight2",
        "price":"nan",
        "taste_lb":"nan",
        "taste_st":"nan",
        "taste_ds":"nan",
        "taste_sa":"nan",
        "taste0_notes":"nan",
        "taste0_top1":"nan",
        "taste0_top2":"nan",
        "taste0_top3":"nan",
        "taste0_notes_num":"nan",
        "taste1_notes":"nan",
        "taste1_top1":"nan",
        "taste1_top2":"nan",
        "taste1_top3":"nan",
        "taste1_notes_num":"nan",
        "taste2_notes":"nan",
        "taste2_top1":"nan",
        "taste2_top2":"nan",
        "taste2_top3":"nan",
        "taste2_notes_num":"nan"
    }


    try:
        driver.get("https://www.vivino.com/larrivet-haut-brion-pessac-leognan/w/81907?year=1998&price_id=47835")#用第14的指引的連結去試

        # 選擇語言 
        language_button = driver.find_elements(By.CSS_SELECTOR, ".simpleLabel_selectedKey__ubPnI")#先定位 “.”的意思是網頁的class的意思 find_elements有s就代表抓的東西會變成list
        language_button[1].click()#所以0是送貨 1是語言
        time.sleep(6)

        #選英文語言
        language_switch = driver.find_element(By.XPATH,"//div[@class='navigationItem_closeWrapper__n28xG']//li[3]")#一樣去定位 這裡用xpath的方法 第三個li的標籤
        language_switch.click()
        time.sleep(5)

        #winery
        winery = driver.find_element(By.CSS_SELECTOR,"a.winery")
        winery = winery.text
        df.at[0,'winery'] = winery
        jsonObject['winery'] = winery

        #wine_name
        wine = driver.find_element(By.CSS_SELECTOR, "span.vintage")#span底下的class 所以也要加“.”vintange
        wine_name = wine.text[:-4]
        df.at[0, 'wine_name'] = wine_name
        jsonObject['wine_name'] = wine_name

        #no
        link = str(df['Link'][0]) #0要改成i
        no = re.findall('.*w/(.*)year.*',link)#?辨識不出來
        no = str(no[0]).replace('?','')
        df.at[0, 'no'] = no
        jsonObject['no'] = no

        #price_id
        price_id = re.findall('.*price_id=(.*).*',link)
        price_id = str(price_id[0])
        df.at[0, 'price_id'] = price_id
        jsonObject['price_id'] = price_id

        #year
        year = wine.text[-4:]
        df.at[0,'year'] = year
        jsonObject['year'] = year

        #country
        origin_info = driver.find_elements(By.CSS_SELECTOR,"a.anchor_anchor__m8Qi-.breadCrumbs__link--1TY6b")
        country = origin_info[0].text
        df.at[0,'country'] = country
        jsonObject['country'] = country

        #region_
        region = origin_info[2].text
        df.at[0,'region'] = region
        jsonObject['region'] = region

        #wine_type
        wine_type = origin_info[3].text
        df.at[0,'wine_type'] = wine_type
        jsonObject['wine_type'] = wine_type

        #grap_id
        grape_id = origin_info[4].text
        df.at[0,'grape_id'] = grape_id
        jsonObject['grape_id'] = grape_id

        #rating
        rating = driver.find_element(By.CSS_SELECTOR, ".vivinoRating_averageValue__uDdPM")
        rating = rating.text
        df.at[0,'rating'] = rating
        jsonObject['rating'] = rating

        #no_ratings
        no_ratings = driver.find_element(By.CSS_SELECTOR, ".vivinoRating_caption__xL84P")
        no_ratings = no_ratings.text[:-8]
        df.at[0,'no_ratings'] = no_ratings
        jsonObject['no_ratings'] = no_ratings

        #price
        price = driver.find_element(By.CSS_SELECTOR, "span.purchaseAvailabilityPPC__amount--2_4GT")
        price = price.text.replace('$','').replace(',','')
        df.at[0,'price'] = price
        jsonObject['price'] = price

        time.sleep(7)

        #highlights
        highlights = driver.find_elements(By.CSS_SELECTOR, ".rightColumn")
        if len(highlights) > 1:
            df.at[0, 'highlight1'] = highlights[0].text
            df.at[0, 'hightlight2'] = highlights[1].text
            jsonObject['highlight1'] = highlights[0].text
            jsonObject['hightlight2'] = highlights[1].text
        elif len(highlights) == 1:
            df.at[0, 'highlight1'] = highlights[0].text
            df.at[0, 'hightlight2'] = "No highlight2"
            jsonObject['highlight1'] = highlights[0].text
            jsonObject['hightlight2'] = "No highlight2"
        else:
            df.at[0, 'highlight1'] = "No highlight1"
            df.at[0, 'hightlight2'] = "No highlight2"
            jsonObject['highlight1'] = "No highlight1"
            jsonObject['hightlight2'] = "No highlight2" 

        driver.execute_script("window.scrollBy(0, 855);")
        time.sleep(3)
        
        #taste
        taste = driver.find_elements(By.CSS_SELECTOR, "span.indicatorBar__progress--3aXLX")
        taste_percent_list = []
        for t in range(4):
            taste_attribute = taste[t].get_attribute('style')
            taste_percent = re.findall('.*left: (.*);.*',taste_attribute)
            taste_percent_list.append(taste_percent)
        taste_percent_list = [str(x) for item in taste_percent_list for x in item]#因為要去掉[[],[],[]]次級的中括號
        taste_lb = taste_percent_list[0]
        taste_st = taste_percent_list[1]
        taste_ds = taste_percent_list[2]
        taste_sa = taste_percent_list[3]
        df.at[0,'taste_lb'] = taste_lb
        df.at[0,'taste_st'] = taste_st
        df.at[0,'taste_ds'] = taste_ds
        df.at[0,'taste_sa'] = taste_sa
        jsonObject['taste_lb'] = taste_lb
        jsonObject['taste_st'] = taste_st
        jsonObject['taste_ds'] = taste_ds
        jsonObject['taste_sa'] = taste_sa

        #taste_notes
        taste_notes = driver.find_elements(By.CSS_SELECTOR, "span.tasteNote__flavorGroup--1Uaen")
        taste0_notes = taste_notes[0].text
        taste1_notes = taste_notes[1].text
        taste2_notes = taste_notes[2].text
        df.at[0,'taste0_notes'] = taste0_notes
        df.at[0,'taste1_notes'] = taste1_notes
        df.at[0,'taste2_notes'] = taste2_notes
        jsonObject['taste0_notes'] = taste0_notes
        jsonObject['taste1_notes'] = taste1_notes
        jsonObject['taste2_notes'] = taste2_notes

        #taste_notes_num
        taste_notes_num = driver.find_elements(By.CSS_SELECTOR, ".tasteNote__mentions--1T_d5")
        taste0_notes_num = str(re.findall('(.*) mentions.*',taste_notes_num[0].text)[0])#因為要去掉[[],[],[]]次級的中括號
        taste1_notes_num = str(re.findall('(.*) mentions.*',taste_notes_num[1].text)[0])
        taste2_notes_num = str(re.findall('(.*) mentions.*',taste_notes_num[2].text)[0])
        df.at[0,'taste0_notes_num'] = taste0_notes_num
        df.at[0,'taste1_notes_num'] = taste1_notes_num
        df.at[0,'taste2_notes_num'] = taste2_notes_num
        jsonObject['taste0_notes_num'] = taste0_notes_num
        jsonObject['taste1_notes_num'] = taste1_notes_num
        jsonObject['taste2_notes_num'] = taste2_notes_num

        #taste_top
        taste_top = driver.find_elements(By.CSS_SELECTOR, ".tasteNote__popularKeywords--1gIa2")
        taste0_top1 = str(taste_top[0].text).split(', ')[0]#(oak, )取list的第一個 用str.split，與re的findall一樣也會回傳一個list
        taste0_top2 = str(taste_top[0].text).split(', ')[1]#(, tobacco, )取第二個
        taste0_top3 = str(taste_top[0].text).split(', ')[2]#(, vanilla) 取第三個
        df.at[0,'taste0_top1'] = taste0_top1
        df.at[0,'taste0_top2'] = taste0_top2
        df.at[0,'taste0_top3'] = taste0_top3
        jsonObject['taste0_top1'] = taste0_top1
        jsonObject['taste0_top2'] = taste0_top2
        jsonObject['taste0_top3'] = taste0_top3

        taste1_top1 = str(taste_top[1].text).split(', ')[0]
        taste1_top2 = str(taste_top[1].text).split(', ')[1]
        taste1_top3 = str(taste_top[1].text).split(', ')[2]
        df.at[0,'taste1_top1'] = taste1_top1
        df.at[0,'taste1_top2'] = taste1_top2
        df.at[0,'taste1_top3'] = taste1_top3
        jsonObject['taste1_top1'] = taste1_top1
        jsonObject['taste1_top2'] = taste1_top2
        jsonObject['taste1_top3'] = taste1_top3

        taste2_top1 = str(taste_top[2].text).split(', ')[0]
        taste2_top2 = str(taste_top[2].text).split(', ')[1]
        taste2_top3 = str(taste_top[2].text).split(', ')[2]
        df.at[0,'taste2_top1'] = taste2_top1
        df.at[0,'taste2_top2'] = taste2_top2
        df.at[0,'taste2_top3'] = taste2_top3
        jsonObject['taste2_top1'] = taste2_top1
        jsonObject['taste2_top2'] = taste2_top2
        jsonObject['taste2_top3'] = taste2_top3
        try:
            with open(filename,'r') as file:#寫入新的元素之前要read
                data = json.load(file)
            with open (filename, 'w') as file:#read完之後才可以write
                data.append(jsonObject)
                file.write(json.dumps(data,indent=4))
            file.close()
        except:
            with open(filename, "w") as file:
                json.dump([jsonObject], file,indent=4)
            file.close()
    except:
        print("error")

    return df

```
